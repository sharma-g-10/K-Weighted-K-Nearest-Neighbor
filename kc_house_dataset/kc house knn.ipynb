```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('kc_house_data.csv')
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df.drop(['id'] , axis = 1 , inplace = True)
```

```python
df.drop_duplicates(keep = 'first',inplace = True)
```

```python
df.drop(['date'] , axis = 1 , inplace = True)
```

```python
df.head()
```

```python
df.corr()
```

```python
df = df[['bedrooms','bathrooms','sqft_living','view','grade','sqft_above','sqft_basement','lat','sqft_living15' ,'price']]
df.head()
```

```python
df.shape
```

```python
X = df.iloc[:,:-1].values
y = df.iloc[:,-1].values
```

```python
X.shape
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

```python
y_train.shape
```

```python
y_train = y_train.reshape(len(y_train),1)
y_test = y_test.reshape(len(y_test),1)

from sklearn.preprocessing import StandardScaler
sc_X = StandardScaler()
sc_y = StandardScaler()
X_train = sc_X.fit_transform(X_train)
X_test = sc_X.transform(X_test)
y_train = sc_y.fit_transform(y_train)
```

```python
from sklearn.neighbors import KNeighborsRegressor
regressor = KNeighborsRegressor(n_neighbors = 18)
regressor.fit(X_train, y_train)
```

```python
y_pred = sc_y.inverse_transform(regressor.predict(X_test))
```

```python
a = np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_test),1)),1)
```

```python
from sklearn.metrics import r2_score
r2_score(y_test, y_pred)
```

```python
from sklearn.metrics import mean_squared_error
mean_squared_error(y_test, y_pred)
```
